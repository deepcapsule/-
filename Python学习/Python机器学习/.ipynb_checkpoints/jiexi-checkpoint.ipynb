```python
import pprint
import numpy as np
def tenOfFile(aheadN,inputfile):
    outList=[]
    with open(inputfile) as infile:
        for line in infile.readlines():
            line = line.split(',')
            lineList=[]
            if len(line) >= aheadN:
                aheadLine = line[0:aheadN]
                for strLine in aheadLine:
                    strLine = strLine.split('|')
                    strLine = list(map(float,strLine))
                    lineList.append(strLine)
                outList.append(lineList)
            
       
    return outList

def charOfFile(inlist):
    outList=[]
    for listone in inlist :
        array2D = np.matrix(listone)
        frameMean = np.mean(array2D[:,1])
        frameMax = np.max(array2D[:,1])
        frameMin = np.min(array2D[:,1])
        frameVar = np.var(array2D[:,1])
        
        updown = array2D[:,0].flatten().A[0]
        timesList = array2D[:,3].flatten().A[0] #.tolist()
        upTimesList = timesList[updown==0].tolist()
        downTimesList = timesList[updown==1].tolist()
        
        
        def mminus(abc):
            bcd=[]
            for i in range(len(abc)-1):
                bcd.append(abc[i+1]*100-abc[i]*100)
            return bcd
        
        if(len(upTimesList)>1):
            upTimeList = mminus(sorted(upTimesList))
            upTimeMean = np.mean(upTimeList)
            upTimeVar = np.var(upTimeList)
        else:
            upTimeMean = 0.0
            upTimeVar = 0.0
        if(len(downTimesList)>1):
            downTimeList = mminus(sorted(downTimesList))
            downTimeMean = np.mean(downTimeList)
            downTimeVar = np.var(downTimeList)
        else:
            downTimeMean = 0.0
            downTimeVar = 0.0
        
        if (upTimeMean==0 and downTimeMean ==0):
            print(updown,upTimesList,downTimesList)
        pushSum = np.sum(array2D[:,4])
        outList.append(list((frameMean,frameMax,frameMin,\
                            frameVar,downTimeMean,downTimeVar,\
                             upTimeMean,upTimeVar,pushSum))) 
        
    return outList
    

inputF = 'mail.txt'
midfile = tenOfFile(10,inputF)
outfile = charOfFile(midfile)

outfile = np.array(outfile)
# np.savetxt('not_mailout.txt',outfile,fmt='%.4f',delimiter='\t')

# pprint.pprint(len(outfile))
```
