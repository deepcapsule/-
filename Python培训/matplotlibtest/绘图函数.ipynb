### 对数坐标图

semilogx():X轴为对数坐标  
semilogy():Y轴为对数坐标  
loglog()  :两个轴都为对数坐标  
其中使用双对数坐标系表示的频率响应曲线通常被成为波特图。

```python
import numpy as np
import matplotlib.pyplot as plt

w = np.linspace(0.1,1000,1000)
p = np.abs(1/(1+0.1j*w))

plt.subplot(221)
plt.plot(w,p,linewidth=2)
plt.ylim(0,1.5)

plt.subplot(222)
plt.semilogx(w,p,linewidth=2)
plt.ylim(0,1.5)

plt.subplot(223)
plt.semilogy(w,p,linewidth=2)
plt.ylim(0,1.5)

plt.subplot(224)
plt.loglog(w,p,linewidth=2)
plt.ylim(0,1.5)

plt.show()
```

### 极坐标图

调用subplot（）创建子图时通过这只polar参数为True，创建一个极坐标子图。

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.arange(0,2*np.pi,0.02)
plt.subplot(121,polar=True)
plt.plot(theta,1.6*np.ones_like(theta),linewidth=2)
plt.plot(3*theta,theta/3,'--',linewidth=2)

plt.subplot(122,polar=True)
plt.plot(theta,1.4*np.cos(5*theta),'--',linewidth=2)
plt.plot(theta,1.8*np.cos(4*theta),linewidth=2)
plt.rgrids(np.arange(0.5,2,0.5),angle=45)
plt.thetagrids([0,45])

plt.show()
```

###  散列图

使用plot（）绘图时，如果指定样式参数为仅绘制数据点（o），那么绘制出来的就是一副散列图，但这种点无法单独指定颜色和大小。  
因此常用scatter（）绘制散列图

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
x = np.random.random(100)
y = np.random.random(100)
'''scatter()前两个参数是数组，分贝指定每个点的X轴和Y轴的坐标。s参数指定点的大小，值和点的面积成正比。它可以是一个数，指定所有点的大小，
也可以是一个数组，分别对每个点指定大小。c参数指定每个点的颜色，可以是数值或数组。这里使用一维数组为每个点指定了一个数值。通过颜色映射表，
每个数值都会与一个颜色对应。默认颜色对应表中蓝色与最小值对应，红色与最大值对应。dangc参数形状为（N,3)或(N,4)的二维数组时，则直接表示每个点
的RGB颜色。marker参数设置点的形状，可以是个表示形状的字符串，也可以是表示多边形的两个元素的元组，第一个元素表示多边形的边数，第二个元素表示
多边形的样式，取值范围为0、1、2、3。0表示多边形，1表示星形，2表示放射形，3表示忽略边数而显示为圆形。alpha设置点的透明度。'''

plt.scatter(x,y,s=x*1000,c=y,marker=(4,1),alpha=0.8,lw=2,facecolors="red")
plt.xlim(0,1)
plt.ylim(0,1)

plt.show()
```

### 等值线图

用等值线绘制二元函数$f(x)=xe^{-x^2-y^2}$

```python
import numpy as np
import matplotlib.pyplot as plt

y,x = np.ogrid[-2:2:200j,-3:3:300j]
z = x*np.exp(-x**2-y**2)

extent = [np.min(x),np.max(x),np.min(y),np.max(y)]

plt.figure(figsize=(12,4))
plt.subplot(121)
cs = plt.contour(z,10,extent=extent)
plt.clabel(cs)  #为等值线标上数值
plt.subplot(122)
plt.contourf(x.reshape(-1),y.reshape(-1),z,20)   #带填充效果的等值线
plt.show()
```

绘制隐函数曲线 $f(x,y)=(x^2+y^2)^4-(x^2-y^2)^2$在f(x,y)=0和f(x,y)-0.1=0时的曲线。

```python
import numpy as np
import matplotlib.pyplot as plt

y,x = np.ogrid[-1.5:1.5:200j,-1.5:1.5:200j]
f = (x**2+y**2)**4-(x**2-y**2)**2

plt.figure(figsize=(9,4))
plt.subplot(121)
extent = [np.min(x),np.max(x),np.min(y),np.max(y)]
#levels设定绘制等值线对应的函数值
cs = plt.contour(f,extent=extent,levels=[0,0.1],colors=['b','r'],linestyles=['solid','dashed'],linewidth=[2,2])

plt.subplot(122)
for c in cs.collections:
    data = c.get_paths()[0].vertices
    plt.plot(data[:,0],data[:,1],color=c.get_color()[0],linewidth=c.get_linewidth()[0])
    
plt.show()
```

### 三维绘图

```python
import numpy as np
import mpl_toolkits.mplot3d
import matplotlib.pyplot as plt

x,y = np.mgrid[-2:2:20j,-2:2:20j]
z = x*np.exp(-x**2-y**2)

ax = plt.subplot(111,projection='3d')
ax.plot_surface(x,y,z,rstride=2,cstride=1,cmap=plt.cm.Blues_r)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)

ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap='hot')
plt.show()
```
