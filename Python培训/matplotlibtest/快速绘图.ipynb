```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,1000)
y = np.sin(x)
z = np.cos(x**2)

#figure创建绘图对象，并使他成为当前绘图对象。
#figsize指定绘图对象的宽度和高度，单位为英寸；
#dpi指定绘图对象的分辨率，即每英寸多少个像素，缺省值为80
plt.figure(figsize=(8,4))  

plt.plot(x,y,label='$sin(x)$',color='red',linewidth=2)
plt.plot(x,z,'b--',label='$cos(x^2)$')          #'b--'指定曲线颜色和类型

plt.xlabel('Times(s)')
plt.ylabel('Volt')
plt.title('PyPlot First Example')
plt.ylim(-1.2,1.2)
plt.legend()  #显示图示
#保存当前图标为图像文件，图像格式由扩展名决定。通过dpi参数指定图像分辨率
plt.savefig('test.png',dpi=120)
plt.show() #显示绘图对象


```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(1)  #创建图表1
plt.figure(2)  #创建图表2
ax1 = plt.subplot(223)  #在图表2中，将其划分为2*2=4个子图并选择第3个，即左下部分
ax2 = plt.subplot(211)  #将图表2重新划分为2*1=2个上下两行的子图，并选择第一个，即第一行，，（此时不能选择第二个，因为左下部分已经被分到ax1）

x = np.linspace(0,3,100)

for i in range(5):
    plt.figure(1)   #选择图表1
    plt.plot(x,np.exp(i*x/3))
    plt.sca(ax1)    #切换到                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                    图表2的子图1
    plt.plot(x,np.sin(i*x))
    plt.sca(ax2)
    plt.plot(x,np.cos(i*x))
plt.show()
```
