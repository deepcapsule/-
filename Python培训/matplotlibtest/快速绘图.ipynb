```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,1000)
y = np.sin(x)
z = np.cos(x**2)

#figure
#figsize
#dpi80
plt.figure(figsize=(8,4))  

plt.plot(x,y,label='$sin(x)$',color='red',linewidth=2)
plt.plot(x,z,'b--',label='$cos(x^2)$')          #'b--'

plt.xlabel('Times(s)')
plt.ylabel('Volt')
plt.title('PyPlot First Example')
plt.ylim(-1.2,1.2)
plt.legend()  #
#dpi
plt.savefig('test.png',dpi=120)
plt.show() #


```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(1)  #1
plt.figure(2)  #2
ax1 = plt.subplot(223)  #22*2=43
ax2 = plt.subplot(211)  #22*1=2ax1

x = np.linspace(0,3,100)

for i in range(5):
    plt.figure(1)   #1
    plt.plot(x,np.exp(i*x/3))
    plt.sca(ax1)    #                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                    21
    plt.plot(x,np.sin(i*x))
    plt.sca(ax2)
    plt.plot(x,np.cos(i*x))
plt.show()
```
