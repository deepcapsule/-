## 优美的蝴蝶曲线

蝴蝶曲线的函数表达式  
$x=\sin(t)\Big(e^{\cos(t)}-2\cos(4t)-\sin^5\big(\dfrac{t}{12}\big)\Big)$  
$y=\cos(t)\Big(e^{\cos(t)}-2\cos(4t)-\sin^5\big(\dfrac{t}{12}\big)\Big)$  
$0\leq5\leq12\pi$

也可以用球坐标表示  
$r=e^{\sin\theta}-2\cos(4\theta)+\sin^5\Big(\dfrac {2\theta-\pi}{24}\Big)$

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0,12*np.pi,0.01)
x = np.sin(t)*(np.e**np.cos(t)-2*np.cos(4*t)-np.sin(t/12)**5)
y = np.cos(t)*(np.e**np.cos(t)-2*np.cos(4*t)-np.sin(t/12)**5)

plt.figure(figsize=(8,6))
plt.axis('off')
plt.plot(x,y,color='blue',linewidth='2')
plt.show()
```
