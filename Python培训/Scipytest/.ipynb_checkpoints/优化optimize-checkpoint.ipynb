用最小二乘法拟合直线

```python
import numpy as np 
from scipy.optimize import leastsq

X = np.array([8.19,2.27,6.39,8.71,4.7,2.66,3.78])
Y = np.array([7.01,2.78,6.47,6.71,4.1,4.23,4.05])

def residuals(p):
    '计算以p为参数的直线和原始数据之间的误差'
    k,b = p
    return Y - (k*X+b)
#leastsq使得residuals()的输出数组的平方和最小，参数初始值为[1.0]
r = leastsq(residuals,[1,0])
k,b = r[0]
print('k=',k,'b=',b)
```

使用最小二乘法对带噪声的正弦波数据进行拟合

```python
import numpy as np 
from scipy.optimize import leastsq

def func(x,p):
    '数据拟合所用的函数A*sin(2*pi*k*x+theta)'
    A,k,theta = p
    return A*2*np.sin(2*np.pi*k*x+theta)

def residuals(p,y,x):
    return y - func(x,p)
x = np.linspace(-2*np.pi,0,1000)
A,k,theta = 10,0.34,np.pi/6  #真实数据的函数参数
y0 = func(x,[A,k,theta])

#加入噪声后的实验数据
y1 = y0 + np.random.randn(len(x))
p0 = [7,0.2,0]  #第一次猜测的函数拟合参数
plsq = leastsq(residuals,p0,args=(y1,x))

print('真实参数：',[A,k,theta])
print('拟合参数：',plsq[0])

import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus']=False

plt.figure(figsize=(10,5))
plt.plot(x,y0,label='真实数据',color='red',linewidth=5)
plt.plot(x,y1,label='带噪声的数据',color='green')
plt.plot(x,func(x,plsq[0]),label='拟合数据',color='yellow')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('正弦拟合')
plt.legend()
plt.show()
```

函数最小值

```python
#用fmin（）计算函数最小值
import scipy.optimize as opt
import numpy as np
import sys

points=[]
def f(p):
    x,y = p
    z = (1-x)**2+100*(y-x**2)**2
    points,append((x,y,z))
    return z

def fprime(p):
    x,y = p
    dx = -2 + 2*x - 400*x*(y-x**2)
    dy = 200*y - 200*x**2
    return np.array([dx,dy])

init_point=(-2,-2)

try:
    method = sys.argv[1]
except:
    method = 'fmin_bfgs'
    
fmin_func = opt.__dict__[method]
if method in ['fmin','fmin_powell']:
    result = fmin_func(f,init_point)
elif method in ['fmin_cg','fmin_bfgs','fmin_l_bfgs_b','fmin_tnc']:
    result = fmin_func(f,init_point,fprime)
elif method in ['fmin_cobyla']:
    result = fmin_func(f,init_point,[])
else:
    print('fmin function not found')
    sys.exit()
```

```python
from scipy.optimize import fsolve
from math import sin
def f(x):
    x0,x1,x2=x.tolist()
    return [
        5*x1+3,
        4*x0*x0-2*sin(x1*x2),
        x1*x2-1.5
    ]
result = fsolve(f,[1,1,1])
print (result)
print(f(result))
```

```python
from scipy.optimize import fsolve
from math import sin,cos

def j(x):
    x0,x1,x2 = x.tolist()
    return [
        [0,5,0],
        [8*x0,-2*x2*cos(x1*x2),-2*x1*cos(x1*x2)],
        [0,x2,x1]
    ]

result = fsolve(f,[1,1,1],fprime=j)

print(result)
```
